### converting numpy array to tensor

```python
np_array = np.array([
    [1,2,3],
    [4,5,6]
])
tensor_numpy= torch.from_numpy(np_array)
print(f"type of the numpy array: {tensor_numpy.dtype}")
print(f"shape of the numpy array: {tensor_numpy.shape}")
```

```python
zero_tensor = torch.zeros(3,3)
one_tensor = torch.ones(1,6)
empty_tensor = torch.empty(4,4)
# static values
some_value =  torch.full((5,6),88)
```
